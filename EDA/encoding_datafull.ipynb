```python

mappings = {
    "periodo": lambda x: int(x) if pd.notna(x) else np.nan,

    "cole_area_ubicacion": {
        "RURAL": 0,
        "URBANO": 1,
    },
    "cole_bilingue": {
        "N": 0,
        "S": 1,
        "NO REPORTADO": -1,
    },
    "cole_caracter": {
        "TÉCNICO": 1,
        "TÉCNICO/ACADÉMICO": 2,
        "ACADÉMICO": 3,
        "NO APLICA": 4,
        "NO REPORTADO": -1,
    },
    "cole_jornada": {
        "MAÑANA": 1,
        "TARDE": 2,
        "NOCHE": 3,
        "SABATINA": 4,
        "UNICA": 5,
        "COMPLETA": 6,
    },
    "cole_naturaleza": {
        "NO OFICIAL": 0,
        "OFICIAL": 1,
    },
    "estu_genero": {
        "F": 0,
        "M": 1,
    },
    "estu_privado_libertad": {
        "S": 0,
        "N": 1,
    },

    # Ordinal variables
    "fami_cuartoshogar": {
        "Uno": 1,
        "Dos": 2,
        "Tres": 3,
        "Cuatro": 4,
        "Cinco": 5,
        "Seis": 6,
        "Seis o mas": 6,
        "Siete": 7,
        "Ocho": 8,
        "Nueve": 9,
        "Diez o más": 10,
    },
    "fami_educacionmadre": {
        "Ninguno": 0,
        "Primaria incompleta": 1,
        "Primaria completa": 2,
        "Secundaria (Bachillerato) incompleta": 3,
        "Secundaria (Bachillerato) completa": 4,
        "Técnica o tecnológica incompleta": 5,
        "Técnica o tecnológica completa": 6,
        "Educación profesional incompleta": 7,
        "Educación profesional completa": 8,
        "Postgrado": 9,
        "No sabe": -1,
        "No Aplica": -1,
    },
    "fami_educacionpadre": {
        "Ninguno": 0,
        "Primaria incompleta": 1,
        "Primaria completa": 2,
        "Secundaria (Bachillerato) incompleta": 3,
        "Secundaria (Bachillerato) completa": 4,
        "Técnica o tecnológica incompleta": 5,
        "Técnica o tecnológica completa": 6,
        "Educación profesional incompleta": 7,
        "Educación profesional completa": 8,
        "Postgrado": 9,
        "No sabe": -1,
        "No Aplica": -1,
    },
    "fami_estratovivienda": {
        "Sin Estrato": 0,
        "Estrato 1": 1,
        "Estrato 2": 2,
        "Estrato 3": 3,
        "Estrato 4": 4,
        "Estrato 5": 5,
        "Estrato 6": 6,
    },
    "fami_personashogar": {
        "Una": 1,
        "Dos": 2,
        "Tres": 3,
        "Cuatro": 4,
        "Cinco": 5,
        "Seis": 6,
        "Siete": 7,
        "Ocho": 8,
        "Nueve": 9,
        "Diez": 10,
        "Once": 11,
        "1 a 2": 1.5,
        "3 a 4": 3.5,
        "5 a 6": 5.5,
        "7 a 8": 7.5,
        "9 o más": 9.5,
        "Doce o más": 12,
    },

    # Binary / nominal
    "fami_tienecomputador": {
        "No": 0,
        "Si": 1,
    },
    "fami_tieneinternet": {
        "No": 0,
        "Si": 1,
    },
    "fami_tienelavadora": {
        "No": 0,
        "Si": 1,
    },
}

def encode_series(series, mapper):
    if callable(mapper):
        return series.apply(mapper)
    return series.map(mapper)

data_encoded = raw_data4.copy()

for col, mapper in mappings.items():
    data_encoded[col] = encode_series(data_encoded[col], mapper)

data_encoded.head()
```
